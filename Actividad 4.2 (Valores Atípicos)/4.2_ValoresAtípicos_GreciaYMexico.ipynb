**Método de Desviación Estandar**

**| Dataframe Grecia**

```python
#Metodo aplicando desviación estandar en Grecia. Encuentro los valores extremos
y=cuantitativasG
Limite_Superior=y.mean() + 3*y.std()
Limite_Inferior=y.mean() - 3*y.std()
print("----------Limite Superior permitido Grecia---------- \n", Limite_Superior)
print("----------Limite Inferior permitido Grecia---------- \n", Limite_Inferior)
```

```python
#obtenemos los datos y los outliers se convierten en nulos en el DataFrame
dataDE_G =  cuantitativasG[(y <= Limite_Superior) & (y >= Limite_Inferior)]
dataDE_G
```

```python
valores_nulos = dataDE_G.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
dataDE_G_limpio = dataDE_G.copy()
dataDE_G_limpio = dataDE_G_limpio.fillna(round(dataDE_G.mean(),1))
dataDE_G_limpio
```

```python
valores_nulos = dataDE_G_limpio.isnull().sum()
valores_nulos
```

```python
DatosDE_G_Limpios = pd.concat([cualitativasG, dataDE_G_limpio], axis=1)
DatosDE_G_Limpios
```

```python
#Corroboramos que no haya valores nulos en el dataframe final
valores_nulos = DatosDE_G_Limpios.isnull().sum()
valores_nulos
```

```python
DatosDE_G_Limpios.to_csv("Grecia_DesviaciónEstandar_Limpio.csv")
```

**Método de Cuartiles**

**| Grecia**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y = cuantitativasG

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q2
iqr = percentile75 - percentile25

Limite_Superior_iqr=percentile75 + 1.5*iqr
Limite_Inferior_iqr=percentile25 - 1.5*iqr
print("----------Limite Superior permitido Grecia---------- \n", Limite_Superior_iqr)
print("----------Limite Inferior permitido Grecia---------- \n", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del dataframe
data_G_iqr = cuantitativasG[(y <= Limite_Superior_iqr) & (y >= Limite_Inferior_iqr)]
data_G_iqr
```

```python
valores_nulos = data_G_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
data_G_iqr_limpio = data_G_iqr.copy()
data_G_iqr_limpio = data_G_iqr_limpio.fillna(round(data_G_iqr.mean(),1))
data_G_iqr_limpio
```

```python
valores_nulos = data_G_iqr_limpio.isnull().sum()
valores_nulos
```

```python
DatosG_IQR_Limpios = pd.concat([cualitativasG, data_G_iqr_limpio], axis=1)
DatosG_IQR_Limpios
```

```python
valores_nulos = DatosG_IQR_Limpios.isnull().sum()
valores_nulos
```

```python
DatosG_IQR_Limpios.to_csv("Grecia_Cuartiles_Limpio.csv")
```
